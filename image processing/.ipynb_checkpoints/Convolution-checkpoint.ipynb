
**Convolution (Tch chp)**: l mt khi nim trong x l tn hiu s nhm bin i thng tin u vo thng qua mt php tch chp vi b lc  tr v u ra l mt tn hiu mi. Tn hiu ny s lm gim nhng c trng m b lc khng quan tm v ch gi nhng c trng chnh.

**Tch chp 2 chiu**: Tch chp thng dng nht l tch chp 2 chiu c p dng trn ma trn u vo v ma trn b lc 2 chiu. Php tch chp ca mt ma trn $X\in R^{W_1\times H_1}$ vi mtb lc(receptive field) $F\in R^{F\times F}$ l mt ma trn $Y\in R^{W_2\times H_2}$ s tr qua nhng bc sau:

**Tnh tch chp ti 1 im**: Ti v tr u tin trn cng ca ma trn u vo ta s lc ra mt ma trn con $X\in R^{F\times F}$ c kch thc bng vi kch thc ca b lc. Gi tr $y_{11}$ tng ng trn $Y$ l tch chp ca$X$ vi$F$ c tnh nh sau:

$$y_{11} = \sum_{i=1}^F \sum_{j=1}^F x_{ij}f_{ij}$$

Tin hnh trt dc theo ma trn theo chiu t tri qua phi, t trn xung di theobc nhy(stride)Sta s tnh c cc gi tr $y_{ij}$ tip theo. Sau khi qu trnh ny kt thc ta thu c trn vn ma trn $Y$ .



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/btl.jpg', 0)
plt.axis('off')
plt.imshow(img, cmap='gray')
```

```python
def conv2d(img, kernel):
    w, h = img.shape[:2]
    wk, hk = kernel.shape[:2]
    ero = wk // 2 + 1
    conv = np.zeros((w-ero, h-ero))

    for i in np.arange(conv.shape[0]):
        for j in np.arange(conv.shape[1]):
            conv[i,j] = np.sum(img[i:i+wk, j:j+hk] * kernel) # Co the su dung ham tensordot
    
    return conv    
```

```python
kernel = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])

plt.axis('off')
plt.imshow(conv2d(img, kernel))
```

```python
# Kernel theo chieu doc
kernel = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])
filtered = cv2.filter2D(src=img, kernel=kernel, ddepth=-1)
plt.axis('off')
plt.imshow(filtered, cmap='gray')
```

```python
# Kernel theo chieu ngang
kernel = np.array([[1,  1,  1],
                   [0,  0,  0],
                   [-1, -1, -1]])
filtered = cv2.filter2D(src=img, kernel=kernel, ddepth=-1)
plt.axis('off')
plt.imshow(filtered, cmap='gray')
```

```python
# blur image
kernel = np.array([[1, 1, 1],
                   [1, 1, 1],
                   [1, 1, 1]]) / 9
filtered = cv2.filter2D(src=img, kernel=kernel, ddepth=-1)
plt.axis('off')
plt.imshow(filtered, cmap='gray')
```

```python
# sharpening image
# Ci ny kh kh hiu, kiu trong 1 kernel, pixel  gia c trng s cao hn cc pixel lin k, 
# gip ni bt hn cc pixel , thnh ra khi nhn vo s ging nh b gy v nh nhn sng ng hn.
kernel = (np.array([[-1, -1, -1],
                    [-1,  9, -1],
                    [-1, -1, -1]]))
filtered = cv2.filter2D(src=img, kernel=kernel, ddepth=-1)
plt.axis('off')
plt.imshow(filtered, cmap='gray')
```
